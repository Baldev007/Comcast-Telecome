````python
# Making a list of feature which has missing values 

features_with_na=[features for features in df.columns if df[features].isna().sum()>0]
````

````python
features_with_na
````

````python
# Percentage of missing value in each feature

for feature in features_with_na:
    print(feature,round(df[feature].isna().mean(),3),"% of Missing value")
````

###  As we have more missing value we need to compair the relation between feature with Missing value  and dependent variable 

````python
for feature in features_with_na:
    data=df.copy()
    data[feature]=np.where(data[feature].isnull(),1, 0)
    plt.figure(figsize=(5,4))
    data.groupby(feature)['SalePrice'].median().plot.bar(color=['Blue','Orange'])
    plt.title(feature)
    plt.show()
````

#### ``` As per the visualization there is a relation between missing value and the dependent variable,hence we need to fill the missing values ``` 

### Numerical Features

````python
# List of numerical variables 

numerical_feature= [feature for feature in df.columns if df[feature].dtypes != 'object']
print('Number of Numerical variables:',len(numerical_feature))
# visualizing the numerical variables 

df[numerical_feature].head()
````

#### ``` In the numerical variables there are 4  variable contains date time information which may be important for our analysis. So Extracting that information ```  

````python
year_feature=[feature for feature in numerical_feature if 'Yr' in feature or 'Year' in feature]
year_feature
````

````python
df[year_feature].nunique()
````

````python
for f in year_feature:
    data=df.copy()
    data.groupby(f)['SalePrice'].median().plot()
    plt.title(f)
    plt.ylabel('Median House Price')
    plt.xlabel(f)
    plt.xticks(rotation=30)
    plt.show()
````

#### *  ``` As per the Bulit year the visualization shows new building has more price  ```
#### * ``` As per YearRemod vs median sale price If the house is remodification year is less then the price of building is high ```
#### * ``` As per GarageYrBlt, if the build year of garage is less the the price is high ```
#### * ``` Old build year has less price during selling ```

````python
discrete_feature= [feature for feature in numerical_feature if len(df[feature].unique())<20 and feature not in year_feature]
````

````python
for discrete in discrete_feature:
    print(discrete, "value's length is ", len(df[discrete].unique()))
    
````

#### ``` Visualizing the relation between discrete variable and the dependent variable ```

````python
for discrete in discrete_feature:
    data=df.copy()
    data.groupby(discrete)['SalePrice'].median().plot.bar()
    plt.title(discrete)
    plt.xlabel(discrete)
    plt.ylabel('Sale Price in median')
    plt.show()
````
